```julia
function is_wanted(num::BigInt)
    if num > 1000 && num % 1000 == 1
        return true
    end
    return false
end

function get_num()
    i = 1
    while true
        n = BigInt(3)^i
        if is_wanted(n)
            println("Number is: 3^$i = ", n)
            break
        end
        i += 1
    end
end

get_num()
```
