```julia
function is_perfect_natural(num::Int)
    arr = Int[]

    for i in 1:1:num / 2
        if num % i == 0
            push!(arr, i)
        end
    end
    return num == sum(arr)
end

function get_perfect_naturals()
    limit = 10000
    arr = Int[]

    for i in 1:1:limit
        if (is_perfect_natural(i))
            push!(arr, i)
        end
    end
    return arr
end

println(get_perfect_naturals())
```
